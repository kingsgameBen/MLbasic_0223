## 文字分群 - 詩詞 (李白、杜甫、白居易)
- tf: term freqency, 這個詞出現在整篇文章的次數，出現越多次，代表這個詞越能代表整篇文章
- idf: inverse document frequency, 這個詞在我蒐集的全部文章出現過的文章數，出現越多次，代表這個詞是一個慣用詞，重要性下降。
- tf-idf: 衡量這篇文章的關鍵詞是什麼的方法
